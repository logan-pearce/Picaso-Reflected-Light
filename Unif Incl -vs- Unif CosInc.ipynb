```python
n = 25

theta = np.linspace(-np.radians(90),np.radians(90),n)
phi = np.linspace(0,np.radians(90),n)
r = 1
phi, theta = np.meshgrid(phi, theta)
x = r*np.sin(phi)*np.cos(theta)
y = r*np.sin(phi)*np.sin(theta)
z = r*np.cos(phi)


theta = np.linspace(-np.radians(90),np.radians(90),n)
cosi = np.linspace(0,1,n)
incl = np.arccos(cosi)
r = 1
incl, theta = np.meshgrid(incl, theta)
x2 = r*np.sin(incl)*np.cos(theta)
y2 = r*np.sin(incl)*np.sin(theta)
z2 = r*np.cos(incl)



%matplotlib inline
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(121, projection='3d')
ax.scatter(x,y,z,color="k",s=20)
plt.xlabel('x')
plt.ylabel('y')
ax.set_zlabel('z')
ax.view_init(elev=0, azim=0)
plt.title(r'$\phi$ drawn from unif[0,90]')

ax2 = fig.add_subplot(122, projection='3d')
ax2.scatter(x2,y2,z2,color="k",s=20)
plt.xlabel('x')
plt.ylabel('y')
ax2.set_zlabel('z')
ax2.view_init(elev=0, azim=0)
plt.title(r'$\cos \phi$ drawn from unif[0,1]')
plt.savefig('../UnifIncl-vs-UnifCosInc.png',bbox_inches='tight')
```

$$\phi = incl = \cos^{-1}(\frac{z}{r})$$
