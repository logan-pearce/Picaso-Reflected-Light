### Cahoy 2010 eqn 1 for planet/star flux ratio:

```python

# Assume 0.5 geometric albedo:
def FluxRatio(Rp, sep, alpha, Ag = 0.5):
    angleterm = (np.sin(alpha) + (np.pi - alpha)*np.cos(alpha)) / np.pi
    Rp = Rp.to(u.km)
    sep = sep.to(u.km)
    C = (2/3) * Ag * ((Rp / sep)**2) * angleterm
    return C
    
FluxRatio(1*u.Rearth, 1*u.au, 0)
```

```python
p['Flux Contrast alpha=0 Ag=0.5'] = np.nan

# Assume full phase:
alpha = 0

for i in range(len(p)):
    if not np.isnan(p.loc[i]['pl_rade']):
        R = p.loc[i]['pl_rade']
    elif not np.isnan(p['M2R infered radius [Rearth]'][i]):
        R = p['M2R infered radius [Rearth]'][i]
    else:
        pass
    sep = p.loc[i]['pl_orbsmax']
    C = FluxRatio(R*u.Rearth, sep*u.au, alpha)
    p['Flux Contrast alpha=0 Ag=0.5'][i] = C
    

# Assume quadrature:
alpha = np.radians(90)

p['Flux Contrast alpha=90 Ag=0.5'] = np.nan

for i in range(len(p)):
    if not np.isnan(p.loc[i]['pl_rade']):
        R = p.loc[i]['pl_rade']
    elif not np.isnan(p['M2R infered radius [Rearth]'][i]):
        R = p['M2R infered radius [Rearth]'][i]
    else:
        pass
    sep = p.loc[i]['pl_orbsmax']
    C = FluxRatio(R*u.Rearth, sep*u.au, alpha)
    p['Flux Contrast alpha=90 Ag=0.5'][i] = C

p
```

```python
def ComputeTeq(StarTeff, StarRad, sep, Ab = 0.3, fprime = 1/4):
    ''' from Seager 2016 Exoplanet Atmospheres eqn 3.9
    https://books.google.com/books?id=XpaYJD7IE20C
    '''
    StarRad = StarRad.to(u.km)
    sep = sep.to(u.km)
    return (StarTeff * np.sqrt(StarRad/sep) * ((fprime * (1 - Ab))**(1/4))).value

ComputeTeq(6000, 1*u.Rsun, 1*u.au)
```

```python
from myastrotools.tools import keplerian_to_cartesian, keplersconstant
things = ['pl_orbincl','pl_orbsmax','pl_orbper','pl_orbeccen','pl_orbtper','pl_orblper','pl_msinie','st_mass',
         'sy_dist']


def GetOrbits(i):
    xs, ys, zs = [],[],[]
    lon = 0
    Decs_mas = np.array([Decs[i]/pl.loc[has_full_orbit_solution[i]]['sy_dist'] 
                        for i in range(len(has_full_orbit_solution))]) * 1000
    meananomarray = np.linspace(0,1,1000) * 2*np.pi
    kep = keplersconstant(pl.loc[has_full_orbit_solution[i]]['st_mass']*u.Msun,Masses[i]*u.Mearth)
    for m in meananomarray:
        pos,vel,acc = keplerian_to_cartesian(
            pl.loc[has_full_orbit_solution[i]]['pl_orbsmax']*u.au,
            pl.loc[has_full_orbit_solution[i]]['pl_orbeccen'],
            pl.loc[has_full_orbit_solution[i]]['pl_orbincl'],
            pl.loc[has_full_orbit_solution[i]]['pl_orblper'],
            lon,
            m,#MeanAnoms[i],
            kep)
        xs.append((pos[0].value/pl.loc[has_full_orbit_solution[i]]['sy_dist'])*1000)
        ys.append((pos[1].value/pl.loc[has_full_orbit_solution[i]]['sy_dist'])*1000)
        zs.append((pos[2].value/pl.loc[has_full_orbit_solution[i]]['sy_dist'])*1000)
    return np.array(xs), np.array(ys), np.array(zs)
```

```python
%matplotlib notebook
import matplotlib as mpl

color= np.linspace(0,1,len(xs))
norm = mpl.colors.Normalize(vmin=0, vmax=1)
plt.figure(figsize=(6,10))

for i in range(len(has_full_orbit_solution)):
    xs, ys, zs = GetOrbits(i)
    
    plt.subplot(4,2,i+1)
    for j in range(len(xs)):
        if zs[j] < 0:
            alpha = 0.05
        else:
            alpha = 1.0
        plt.scatter(ys[j],xs[j],c=color[j],cmap='viridis',norm=mpl.colors.Normalize(vmin=0, vmax=1),
                   s = 3, alpha = alpha)
    plt.axvline(x=0, ls=':', color='grey', alpha = 0.5)
    plt.annotate('Phase = {:.2f}'.format(Alpha[i]), xy = (0.5,0.05), xycoords='axes fraction')
    #cbar = plt.colorbar()
    #cbar.ax.set_ylabel('Orbit Phase')
    plt.scatter(RAs_mas[i],Decs_mas[i], edgecolors='black',facecolors='None',s=100,lw=2)
    plt.scatter(0,0,marker='*',color='orange',s=200)
    plt.gca().set_aspect('equal', adjustable='datalim')
    #plt.xlim(-0.05,0.05)
    #plt.ylim(-0.05,0.05)
    plt.gca().invert_xaxis()
    plt.grid(ls=':')
    plt.title(planets_names[i])
    
#plt.annotate('Note: long of ascending node \n not available in exoplanet archive,\n set to 0 for all planets; \n Line of nodes marked by dashed line',
#            xy = (0.6,0.1), xycoords='figure fraction')
plt.tight_layout()
plt.savefig('GMagAO-X-targets-with-known-phase-on1Jan2035-fullorbits.png', bbox_inches='tight')
```
